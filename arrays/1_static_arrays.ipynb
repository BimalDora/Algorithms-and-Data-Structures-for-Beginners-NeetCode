# Remove Duplicates From Sorted Array

You are given an integer array nums sorted in non-decreasing order. Your task is to remove duplicates from nums in-place so that each element appears only once.

After removing the duplicates, return the number of unique elements, denoted as k, such that the first k elements of nums contain the unique elements.

Note:

The order of the unique elements should remain the same as in the original array.
It is not necessary to consider elements beyond the first k positions of the array.
To be accepted, the first k elements of nums must contain all the unique elements.
Return k as the final result.

Example 1:

Input: nums = [1,1,2,3,4]

Output: [1,2,3,4]
Explanation: You should return k = 4 as we have four unique elements.

Example 2:

Input: nums = [2,10,10,30,30,30]

Output: [2,10,30]
Explanation: You should return k = 3 as we have three unique elements.

Constraints:

1 <= nums.length <= 30,000
-100 <= nums[i] <= 100
nums is sorted in non-decreasing order.

```python
# Brute Force Solution
def remove_duplicates_brute_force(nums : list[int]) -> tuple:
    k = len(nums)
    i = 0
    j = 1
    while k > j:
        if nums[i] == nums[j]:
            k -= 1
            nums.remove(nums[i])
        else:
            i += 1
            j += 1
    return nums, k
```

```python
remove_duplicates_brute_force([2,10,10,30,30,30])
```

```python
def remove_duplicates_optimized(nums: list[int]) -> int:
    """
    Removes duplicates from a sorted array in-place.
    Returns the number of unique elements (k).
    """
    if not nums:
        return 0
        
    write_index = 1
    
    for i in range(1, len(nums)):
        if nums[i] != nums[i-1]:
            nums[write_index] = nums[i]
            write_index += 1
            
    return write_index
```

```python
test_data = [1, 1, 2, 4, 4, 8]
k = remove_duplicates_optimized(test_data)
print(f"Output: {k}, Array: {test_data[:k]}")
```

# Remove Element

You are given an integer array nums and an integer val. Your task is to remove all occurrences of val from nums in-place.

After removing all occurrences of val, return the number of remaining elements, say k, such that the first k elements of nums do not contain val.

Note:

The order of the elements which are not equal to val does not matter.
It is not necessary to consider elements beyond the first k positions of the array.
To be accepted, the first k elements of nums must contain only elements not equal to val.
Return k as the final result.

Example 1:

Input: nums = [1,1,2,3,4], val = 1

Output: [2,3,4]
Explanation: You should return k = 3 as we have 3 elements which are not equal to val = 1.

Example 2:

Input: nums = [0,1,2,2,3,0,4,2], val = 2

Output: [0,1,3,0,4]
Explanation: You should return k = 5 as we have 5 elements which are not equal to val = 2.

Constraints:

0 <= nums.length <= 100
0 <= nums[i] <= 50
0 <= val <= 100

```python
def remove_element(nums: list[int], val: int) -> int:
    """
    Removes all instances of 'val' from nums in-place.
    Returns the new length.
    """
    write_index = 0
    for i in range(len(nums)):
        if nums[i] != val:
            nums[write_index] = nums[i]
            write_index += 1
    return write_index
```

```python

test_data = [3, 2, 2, 3]
val_to_remove = 3
new_len = remove_element(test_data, val_to_remove)
print(f"Length: {new_len}, Array: {test_data[:new_len]}")
```

```python
# Insert an element at specific index
def insert_at_index_brute_force(nums : list[int], val : int, index : int) -> list:
    n = len(nums)
    if index >= n:
        return ["Index out of range"]
    else:
        for i in range(n-1, index, -1):
            nums[i] = nums[i-1]
    nums[index] = val
    return nums
```

```python
insert_at_index_brute_force([1,2,3,6,9,8,5], -99, 3)
```

```python
# Remove an element
def remove_element_brute_force(nums: list[int], val: int) -> list:
    idx = 0
    removed_count = 0
    for ele in nums:
        if ele == val:
            for i in range(idx, len(nums)-1):
                nums[i] = nums[i+1]
            removed_count += 1
        idx += 1
    return nums[0:len(nums)-removed_count]
```

```python
nums = [1,2,1,3,6]
remove_element_brute_force(nums, 1)
```
